```python
import requests
import csv
from bs4 import BeautifulSoup

url = "https://resultados.as.com/resultados/motor/motogp/clasificacion/"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

name = soup.find_all('tr', class_='row-table-datos')

clas = list()

for i in name:
  clas.append(i.text)

for e in range(6):
  clas.pop(-1)


for v in clas:
  cortar = v.rindex(' ')
  print(v[2:v.index(' ',cortar)])
  print('----------------------')







```
